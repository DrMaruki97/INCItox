```python
import requests
import pdfplumber
import re
import pandas as pd
import io

def fetch_pdf(url):
    response = requests.get(url)
    response.raise_for_status()  # Ensure the request was successful
    return response.content

def extract_text_from_pdf(pdf_content):
    text = ""
    with pdfplumber.open(io.BytesIO(pdf_content)) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text
    return text

def capture_surrounding_text(text, search_terms):
    # Define the regex pattern to match the terms followed by a numerical value and unit
    pattern = r'\b(?:{})\b(?:(?!\b(?:{})\b).)*?(\d+\.?\d* ?(?:mg/kg|mg|g/kg|g|µg/kg|µg|kg))'.format('|'.join(search_terms), '|'.join(search_terms))
    
    # Find all matches
    matches = re.finditer(pattern, text, re.IGNORECASE | re.DOTALL)
    
    result_contexts = []
    for match in matches:
        match_text = match.group()
        
        # Split the text into words
        words = text.split()
        
        # Find the starting and ending index of the match in words list
        match_start = text.find(match_text)
        match_end = match_start + len(match_text)
        
        # Get 10 words before and after the match
        start_idx = max(0, len(words[:match_start].split()) - 10)
        end_idx = len(words[:match_end].split()) + 10
        
        context_words = words[start_idx:end_idx]
        context = ' '.join(context_words)
        
        result_contexts.append((match.group(1), context))
    
    return result_contexts

def build_dataframe(contexts):
    data = {'values': [val for val, _ in contexts],
            'context': [ctx for _, ctx in contexts]}
    df = pd.DataFrame(data)
    return df

def main(url):
    # Step 1: Fetch PDF
    pdf_content = fetch_pdf(url)
    
    # Step 2: Extract text from PDF
    text = extract_text_from_pdf(pdf_content)
    
    # Step 3: Define search terms
    search_terms = ["NOAEL", "LD50", "LD 50", "LD"]
    
    # Step 4: Capture surrounding text
    contexts = capture_surrounding_text(text, search_terms)
    
    # Step 5: Build DataFrame
    df = build_dataframe(contexts)
    
    return df



# Example usage
url = 'https://cir-reports.cir-safety.org/view-attachment?id=0334b6a1-8c74-ec11-8943-0022482f06a6'  # Replace with actual PDF URL
df = main(url)
print(df)
```

```python
import requests
import io
import re
from pdfminer.high_level import extract_text
from urllib.parse import urlparse

def download_pdf(url):
    response = requests.get(url)
    return io.BytesIO(response.content)

def extract_text_from_pdf(pdf_file):
    return extract_text(pdf_file)

def find_keywords_and_values(text):
    pattern = r'(.{0,150})(NOAEL|LD50|LD)\s*(\d+(?:[.,]\d+)?)\s*(.{0,150})'
    matches = re.finditer(pattern, text, re.IGNORECASE | re.DOTALL)
    
    results = []
    for match in matches:
        before, keyword, value, after = match.groups()
        results.append({
            'keyword': keyword,
            'value': value,
            'context': before.strip() + keyword + value + after.strip()
        })
    
    return results

def main(url):
    try:
        pdf_file = download_pdf(url)
        text = extract_text_from_pdf(pdf_file)
        results = find_keywords_and_values(text)
        
        if results:
            for result in results:
                print(f"Keyword: {result['keyword']}")
                print(f"Value: {result['value']}")
                print(f"Context: {result['context']}")
                print('-' * 50)
        else:
            print("No matching keywords and values found.")
    
    except Exception as e:
        print(f"An error occurred: {str(e)}")

if __name__ == "__main__":
    pdf_url = 'https://cir-reports.cir-safety.org/view-attachment?id=0334b6a1-8c74-ec11-8943-0022482f06a6'
    if not urlparse(pdf_url).scheme:
        pdf_url = "http://" + pdf_url
    main(pdf_url)
```

```python
def extract_text_from_pdf_url(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            pdf_data = BytesIO(response.content)
            text_pages = []
            document = pdf.PdfReader(pdf_data)
            for page_num in range(len(document.pages)):
                try:
                    page = document.pages[page_num]
                    page_text = page.extract_text()
                    if page_text:
                        text_pages.append((page_text, page_num + 1))
                    
                except Exception as e:
                    pass
            return text_pages
        
    except Exception as e:
        pass

def extract_noael_and_ld50(text_pages):
    noael_pattern = re.compile(r'(.*?NOAEL.*?\d+\.?\d*\s*[a-zA-Z/]+.*?(\.|$))', re.IGNORECASE)
    ld50_pattern = re.compile(r'(.*?LD50.*?\d+\.?\d*\s*[a-zA-Z/]+.*?(\.|$))', re.IGNORECASE)
    
    noael_matches = []
    ld50_matches = []
    
    for text, page_num in text_pages:
        lines = text.split("\n")
        for i, line in enumerate(lines):
            if re.search(noael_pattern, line):
                value = re.match('(\d+(?:[.,]\d+)\s*.g\/Kg[\s*\/.*\/d.*]*)',line,re.IGNORECASE)
                previous_line = lines[i - 1] if i > 0 else ""
                formatted_match = f"{previous_line} {line}"
                noael_matches.append((formatted_match, value))
            if re.search(ld50_pattern, line):
                previous_line = lines[i - 1] if i > 0 else ""
                formatted_match = f"{previous_line} {line}"
                ld50_matches.append((formatted_match, page_num))
    
    return noael_matches, ld50_matches

pages = extract_text_from_pdf_url(url)
noaels,ld50s = extract_noael_and_ld50(pages)
print(noaels)
print(ld50s)
```
